# Gaussian Scale Mixture

## original GSM paper

M. J. Wainwright and E. P. Simoncelli, “Scale Mixtures of Gaussians and the Statistics of Natural Images,” presented at the Advances in Neural Information Processing Systems 12, [NIPS Conference, Denver, Colorado, USA, November 29 - December 4, 1999], 1999, pp. 855–861.

Notes: main contribution of this paper is that it introduces GSM to model natural images (here, modeling the coefficients of some wavelet filter bank).

A lot of details are missing.

For example, in 2.2, no description on how to fit Q is given, although somehow they know Q from Figure 3.

In 2.3, I guess w(s) in the first equation is some white (Gaussian) noise. But anyway, main takeaway is that they construct a generative model, where the scale parameters are more and more indpendent as pixels' distance become large.



Notice that, most histogram shown in the paper are conditional.

Gaussian scale mixture is to move out the heavy-tail (bow-tie) behavior in conditional histogram.

For a 2D Gaussian. if two dims are independent, then conditoinal histogram should give a bar with same width everywhere. But jointly, the contour should be round.

A tutorial is at <http://www.gatsby.ucl.ac.uk/~turner/JournalClubs/GSMs/wainwrightetal.pdf>. [local copy](./_gsm/wainwrightetal.pdf) provided.

Wenhao Zhang's notes on GSM is [here](./gsm/170201-GaussScaleMixMdl.pdf)

## 2 model Mixture GSM

R. C. Cagli, P. Dayan, and O. Schwartz, “Statistical Models of Linear and Nonlinear Contextual Interactions in Early Visual Processing,” presented at the Advances in Neural Information Processing Systems 22: 23rd Annual Conference on Neural Information Processing Systems 2009. Proceedings of a meeting held 7-10 December 2009, Vancouver, British Columbia, Canada., 2009, pp. 369–377.

Notes: this paper uses mixture of gaussian scale mixture to model some neural response. Then they showed how to learn this model. This is probably the first paper I know that handles learning of (Rayleigh prior) GSM (or mixture of GSM)

Using Rayleigh prior, I guess, is primarily making some computation tractable.

## 5 model Mixture GSM

[1]	R. Coen-Cagli, P. Dayan, and O. Schwartz, “Cortical Surround Interactions and Perceptual Salience via Natural Scene Statistics,” PLoS Comput Biol, vol. 8, no. 3, p. e1002405, Mar. 2012.

Notes: a 5-model mixture of GSM.

I don't like it, because the training is so fragile, based on what Jeff Helt found.



the model is defined at pp. 6 (Model equations).

for inference, it's given in Eqs. 2-6. Notice that Eq. 4 is not wrong (sqrt of lambda); check their 2009 NIPS paper (Statistical Models of Linear and Nonlinear Contextual Interactions in Early Visual Processing)






Notice that it's very restrictive. Below is what their five models are. two phases of filters are ignored.

They have 4 types of filters (4 orietatnsions), located in 9 locations in a spatial grid, as shown in Figure 4.

In total, there are 4x9 = 36 filters (or 72 for two phases)

we can group all filters for center location to group0,

all 0 deg filters for surround to group1 (8 in total, or 16 if you count phase).
all 45 deg ones to group2,
all 90 deg ones to group3
all 135 deg ones to group4.

these 5 groups partition the whole set of filters.

They then have 5 models.

model0, 5 groups are modeled as independent GSMs
model1, group0 and group1 are modeled jointly as a GSM, and other three are modeled as independent GSMs again.
model2, group0 and group2 are modeled jointly; others are independent.
and so on.

Effectively this is what Eq. (10) and (11) are saying.

In the code for training, there is some trick, such as adding noise (check MGSM_MULTISUR_EM.m in their code; there are some use of rand. A local version of code is [here](./_gsm/mgsm_multisur_r1.zip).
